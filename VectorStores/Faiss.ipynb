# Overview of FAISS

**FAISS** (Facebook AI Similarity Search) is a library developed by **Facebook AI Research (FAIR)** that enables efficient similarity search and clustering of dense vectors. It is widely used for tasks like nearest neighbor search, especially when dealing with large-scale datasets.

## Key Points about FAISS

1. **Developed by Facebook AI**  
   - Created by Facebook AI Research (FAIR) to solve the problem of finding nearest neighbors in large datasets efficiently.

2. **Optimized for High-Dimensional Data**  
   - FAISS is designed to work with high-dimensional dense vectors, often found in applications like word embeddings or image features.

3. **Supports Both CPU and GPU**  
   - It can run on CPUs and GPUs, making it suitable for both small and large datasets, and capable of leveraging the power of GPUs for faster computations.

4. **Efficient Similarity Search**  
   - FAISS accelerates similarity searches such as k-nearest neighbors (KNN), making it highly efficient for finding related data points in large datasets.

5. **Vector Quantization**  
   - FAISS uses advanced techniques like quantization (IVF, PQ, etc.) to reduce memory usage and speed up the search process.

6. **Large-Scale Search**  
   - FAISS is capable of handling billions of vectors, making it highly scalable for large-scale machine learning tasks.

7. **Available in Multiple Languages**  
   - FAISS is available in both C++ and Python, allowing flexibility in integrating it with various applications.

## Advantages of FAISS

1. **High Performance**  
   - FAISS offers extremely fast search capabilities, especially when using GPU-based indexing and retrieval.

2. **Scalability**  
   - FAISS can scale to datasets with billions of vectors while maintaining efficient search performance.

3. **Versatile Indexing**  
   - It provides different types of indexing algorithms (e.g., flat indexing, inverted file indexing) to suit different use cases based on memory and speed requirements.

4. **Supports Approximate Nearest Neighbor (ANN) Search**  
   - FAISS allows for approximate nearest neighbor searches, offering a balance between speed and accuracy, which is beneficial for real-time applications.

5. **Customizable Precision-Recall Tradeoff**  
   - Users can configure FAISS to adjust between higher precision and faster retrieval times, making it adaptable for different tasks.

6. **Optimized for High-Dimensional Data**  
   - Efficiently handles high-dimensional vector spaces, such as embeddings used in machine learning and NLP models.

7. **Low Memory Usage**  
   - FAISS provides memory-efficient solutions using compression techniques like product quantization (PQ).

8. **Integration with ML Pipelines**  
   - FAISS can be easily integrated into machine learning pipelines for tasks like recommendation systems and document retrieval.

## Use Cases of FAISS

1. **Recommendation Systems**  
   - FAISS can be used to find similar items for recommendation based on user interaction data or content embeddings.

2. **Image and Video Search**  
   - Efficiently finds similar images or videos in large datasets by searching over their feature vectors (e.g., in image recognition tasks).

3. **Natural Language Processing (NLP)**  
   - Used for tasks like finding similar word embeddings or document embeddings for information retrieval or clustering.

4. **Anomaly Detection**  
   - FAISS can help detect outliers by finding nearest neighbors and measuring the similarity of vectors in a dataset.

5. **Clustering in High-Dimensional Spaces**  
   - Enables fast clustering of large datasets by finding nearest neighbors and grouping similar data points efficiently.

6. **Search in Large Vector Databases**

```python
from langchain_community.document_loaders import TextLoader
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import OllamaEmbeddings
from langchain_text_splitters import CharacterTextSplitter
```

```python
loader=TextLoader("speech.txt")
documents=loader.load()
text_splitter=CharacterTextSplitter(chunk_size=500,chunk_overlap=20)
# text_splitter is an instance of the CharacterTextSplitter class.
docs=text_splitter.split_documents(documents)
```

```python
docs
```

```python
embeddings=OllamaEmbeddings(model="gemma:2b")
db=FAISS.from_documents(docs,embeddings)
db
```

```python
#  Querying db
query="What is more precious than peace, what shall we fight for? "
```

```python
docs=db.similarity_search(query)
docs
```

```python
docs[0].page_content
```

A Retriever is a component or tool responsible for fetching relevant documents or data based on a query.

FAISS: When using FAISS for document retrieval, the retriever would use the FAISS index to find the nearest neighbors (most similar documents) to the query vector.
Question Answering Systems: In QA systems, a retriever might fetch relevant passages from a corpus that are used to generate or rank the final answer.

```python
retr=db.as_retriever()
docs=retr.invoke(query)
```

```python
docs[0].page_content
```

#### Similarity Search with score
There are some FAISS specific methods. One of them is similarity_search_with_score, which allows you to return not only the documents but also the distance score of the query to them. The returned distance score is L2 distance. Therefore, a lower score is better.

```python
docs_and_score=db.similarity_search_with_score(query)
docs_and_score
# manhatten distance
```

```python
# To Pass embeddings instead of text
embedding_vector=embeddings.embed_query(query)
embedding_vector
```

```python
docs_score=db.similarity_search_by_vector(embedding_vector)
docs_score
```

```python
newdf=FAISS.load_local("a_b",embeddings,allow_dangerous_deserialization=True)
```
