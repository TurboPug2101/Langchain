```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_huggingface import HuggingFaceEmbeddings
```

 1. Sentence-BERT (SBERT) a modification of the BERT model that uses a Siamese network to generate sentence embeddings
2. SBERT uses a Siamese architecture, which has two identical BERT architectures that share the same weights. During training, SBERT processes two sentences as pairs, concatenating the two embeddings that each BERT outputs

```python
embeddings=HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
```

```python
text="this is my text"
query_result=embeddings.embed_query(text)
query_result
```

```python
len(query_result)
```

```python
doc_result=embeddings.embed_documents([text,"i like running"])
```

```python
len(doc_result)
```
